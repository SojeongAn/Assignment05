```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
from skimage import io, color
from skimage import exposure

file_image	= 'cau.jpg'

im_color 	= io.imread(file_image)
im_gray  	= color.rgb2gray(im_color)

ker 		= np.array([[0,0,0],[0,1,0],[0,0,0]]) 
im_conv		= signal.convolve2d(im_gray, ker, boundary='symm', mode='same') 

p1 = plt.subplot(2,2,1)
p1.set_title('color image')
plt.imshow(im_color)
plt.axis('off')

p2 = plt.subplot(2,2,2)
p2.set_title('gray image')
plt.imshow(im_gray, cmap='gray')
plt.axis('off')

p3 = plt.subplot(2,2,3)
p3.set_title('convolution kernel')
plt.imshow(ker, cmap='gray')
plt.axis('off')

p4 = plt.subplot(2,2,4)
p4.set_title('convolution result')
plt.imshow(im_conv, cmap='gray')
plt.axis('off')

plt.show()
```

# Computing the magnitude of the gradient: $f(x)= \sqrt{(vec1-vec2)^2}$

```python
def computeMagnitude(vec1,vec2): 
    return np.sqrt(vec1**2 + vec2**2)
```

# Computing the direction of the gradient: $\tan^{-1}(x)$

```python
def computeDirection(vec1,vec2):
    return np.arctan(vec2/vec1)
```

# derivation in x-direction

```python
kernelX = np.array([[2,2,1],[0,1,1],[1,1,0]]) #Declare kernelX
derivativeX= signal.convolve2d(im_gray, kernelX, boundary='symm', mode='same') #Convolve

p1 = plt.subplot(2,2,1)
p1.set_title('kernelX')
plt.imshow(kernelX, cmap='gray')
plt.axis('off')

p2 = plt.subplot(2,2,2)
p2.set_title('kernelX result')
plt.imshow(derivativeX, cmap='gray')
plt.axis('off')
```

# derivation in y-direction

```python
kernelY = np.array([[-1,0,0],[0,0,-1],[0,-1,0]]) #Declare kernelY
derivativeY= signal.convolve2d(im_gray, kernelY, boundary='symm', mode='same') #Convolve

p1 = plt.subplot(2,2,1)
p1.set_title('kernelY')
plt.imshow(kernelY, cmap='gray')
plt.axis('off')

p2 = plt.subplot(2,2,2)
p2.set_title('kernelY result')
plt.imshow(derivativeY, cmap='gray')
plt.axis('off')
```

# Absolute value of gradient

```python
gradient=computeMagnitude(derivativeX,derivativeY) #Compute magnitude
plt.imshow(gradient, cmap='gray')
plt.axis('off')
```

# Direction of gradient

```python
gradient2=computeDirection(derivativeX,derivativeY) #Compute direction
plt.imshow(gradient2, cmap='gray')
plt.axis('off')
```

```python
plt.imshow(gradient2, cmap='hsv')
plt.axis('off')
```

```python
smoothKernel = np.array([[1/3,1/3,1/3],[1/3,1/3,1/3],[1/3,1/3,1/3]])
smooth= signal.convolve2d(im_gray, smoothKernel, boundary='symm', mode='same') 

p1 = plt.subplot(1,2,1)
p1.set_title('smoothKernel')
plt.imshow(kernelY, cmap='gray')
plt.axis('off')

p2 = plt.subplot(1,2,2)
p2.set_title('kernelX result')
plt.imshow(smooth, cmap='gray')
plt.axis('off')
```
